#                                 ICC_T20_WC_2022_DataCleaning

```python
# import libraries
import pandas as pd
import json
```

```python
# load the json file
with open('t20_json_files/t20_wc_match_results.json') as f:
    data=json.load(f)

    
#Converts data from json file to a table format using pandas DataFrame    
df_matches = pd.DataFrame(data[0]['matchSummary'])
df_matches.head(3)
```

```python
df_matches.shape
```

```python
df_matches.rename({'scorecard':'Match ID'}, axis =1, inplace = 1) #rename scorecard to Match ID 
df_matches.head(10)
```

```python
# transfer the dataframe to csv file
df_matches.to_csv('t20_csv_files/match_results.csv',index=True)
```

```python
match_ids_dict = {}

for index, row in df_matches.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['Match ID']
    match_ids_dict[key2] = row['Match ID']

match_ids_dict
```

Batting summary

```python
# load the json file
with open('t20_json_files/t20_wc_batting_summary.json') as f:
    data1=json.load(f)

    # loads all batting summary of all matches into one single table by joining batting summary of every match using extend
    all_records = []
    for i in data1:
        all_records.extend(i['battingSummary'])    

df_bat = pd.DataFrame(all_records)
df_bat
```

```python
# Adds column 'Out/Not out' 
df_bat["Out/Not_out"] = df_bat.dismissal.apply(lambda x: 'Out' if len(x)>0 else 'Not out')
# remove the dismissal column
df_bat.drop(columns = ['dismissal'], inplace = True)
df_bat.head(10)
```

```python
df_bat["batsmanName"] = df_bat['batsmanName'].replace('â€','')
df_bat["batsmanName"] = df_bat['batsmanName'].replace('\xa0','')
df_bat.head(10)
```

```python
# adds match ID to batting table by linking 'Match ID' column to 'match' column using map feature
df_bat['Match ID']= df_bat['match'].map(match_ids_dict) 
df_bat.head(10)
```

```python
# transfer the dataframe to csv file
df_bat.to_csv('t20_csv_files/batting_summary.csv',index=True)
```

Bowling Summary

```python
# load the json file
with open('t20_json_files/t20_wc_bowling_summary.json') as f:
    data2=json.load(f)
     # loads all bowling summary of all matches into one single table by joining bowling summary of every match using extend
    bowl_records=[]
    for i in data2:
        bowl_records.extend(i['bowlingSummary'])
        
df_bowl=pd.DataFrame(bowl_records)
df_bowl
```

```python
# adds match ID to batting table by linking 'Match ID' column to 'match' column using map feature
df_bowl['Match ID']= df_bowl['match'].map(match_ids_dict) 
df_bowl
```

```python
# transfer into a csv file
df_bowl.to_csv('t20_csv_files/bowling_summary.csv')
```

Players Information

```python
# load the json file
with open('t20_json_files/t20_wc_player_info.json') as f:
    data3=json.load(f)
        
df_players_info=pd.DataFrame(data3)
df_players_info.head(10)
```

```python
df_players_info['name'] = df_players_info['name'].apply(lambda x: x.replace('â€', ''))
df_players_info['name'] = df_players_info['name'].apply(lambda x: x.replace('†', ''))
df_players_info['name'] = df_players_info['name'].apply(lambda x: x.replace('\xa0', ''))
df_players_info.head(10)
```

```python
df_players_info.to_csv('t20_csv_files/player_info.csv')
```
